```python
car_sales_missing =  pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing["Odometer"].mean()
```

```python
 car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].fillna(
    car_sales["Odometer (KM)"].mean()
)
```

```python
car_sales
```

```python
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
car_sales_missing_dropped.to_csv("car_sales_missing_dropped.csv")
```

```python
# Column from series
seats_column = pd.Series([5,5,5,5,5])

# New column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
# bina iske true condition run nhi karegi
print(car_sales.head())
```

```python
#column from python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.4, 5.6, 8.5, 7.9, 9.9, 8.6]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
car_sales["Seats"] = car_sales["Seats"].fillna(
    car_sales["Seats"].mean()
)
```

```python
car_sales
```

```python
car_sales["Total fuel used(L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```

```python
car_sales
```

```python
# Create a coloumn from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales.dtypes
```
